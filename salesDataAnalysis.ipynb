```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sales.csv")
print(df.head())
```

```python
print("üí∞ Total Sales:", df['Sales'].sum())
```

```python
product_sales = df.groupby("Product")["Sales"].sum()
print("\nüì¶ Sales by Product:\n", product_sales)
```

```python
region_sales = df.groupby("Region")["Sales"].sum()
print("\nüåç Sales by Region:\n", region_sales)
```

```python
product_sales.plot(kind="bar", title="Sales by Product", ylabel="Total Sales",xlabel="product",color="skyblue")
plt.show()
```

```python
region_sales.plot(kind="pie", title="Sales by Region", autopct='%1.1f%%')
plt.show()
```
